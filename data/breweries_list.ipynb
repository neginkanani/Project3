```python
import pandas as pd
```

```python
untapped_micros_us_brewery = pd.read_csv('untapped_micros_us_brewery.csv')
untapped_micros_us_brewery_df = untapped_micros_us_brewery.drop(columns=['Unnamed: 0'])
#untapped_micros_us_brewery_df
```

```python
untapped_contract_brewery_df = pd.read_csv("untapped_contract_brewery.csv")
untapped_contract_brewery_df = untapped_contract_brewery_df.drop(columns=['Unnamed: 0'])
#untapped_meadery_df
```

```python
untapped_meadery_df = pd.read_csv("untapped_meadery.csv")
untapped_meadery_df = untapped_meadery_df.drop(columns=['Unnamed: 0'])
#untapped_meadery_df
```

```python
untapped_cidery_df = pd.read_csv("untapped_cidery.csv")
untapped_cidery_df = untapped_cidery_df.drop(columns=['Unnamed: 0'])
#untapped_cidery_df
```

```python
untapped_brew_pub_df = pd.read_csv('untapped_brew_pub.csv') 
untapped_brew_pub_df = untapped_brew_pub_df.drop(columns=['Unnamed: 0'])
#untapped_brew_pub_df
```

```python
brewery_untapped_nano_df = pd.read_csv('untapped_nano.csv') 
brewery_untapped_nano_df = brewery_untapped_nano_df.drop(columns=['Unnamed: 0'])
#brewery_untapped_nano_df
```

```python
untapped_macros_us_brewery = pd.read_csv("untapped_macros_us_brewery.csv")
untapped_macros_us_brewery_df = untapped_macros_us_brewery.drop(columns=['Unnamed: 0'])
#untapped_macros_us_brewery_df 
```

```python
untapped_macros = pd.read_csv("untapped_macros.csv")
brewery_untapped_macros_df = untapped_macros.drop(columns=['Unnamed: 0'])
#untapped_macros_us_brewery_df 
```

```python
brewery_untapped_regional_df = pd.read_csv('untapped_regional.csv') 
brewery_untapped_regional_df = brewery_untapped_regional_df.drop(columns=['Unnamed: 0'])
#brewery_untapped_regional_df
```

```python
brewery_untapped_micro_df = pd.read_csv('untapped_micros.csv') 
brewery_untapped_micro_df = brewery_untapped_micro_df.drop(columns=['Unnamed: 0'])

#brewery_untapped_micro_df
```

```python
#untapped_macros_us_brewery_df, 
dfs = [brewery_untapped_micro_df, 
       untapped_micros_us_brewery_df, 
       untapped_contract_brewery_df, 
       untapped_meadery_df, 
       untapped_cidery_df, 
       untapped_brew_pub_df, 
       brewery_untapped_nano_df, 
       untapped_macros_us_brewery_df, 
       brewery_untapped_macros_df, 
       brewery_untapped_regional_df, 
       brewery_untapped_micro_df]

brewery_untapped_all = pd.concat(dfs)
```

```python
brewery_untapped_all[['number_beers','b', 'number_ratings','r']] = brewery_untapped_all.number_beers.str.split(' ', expand=True)
brewery_untapped_all = brewery_untapped_all.drop(columns=['b','r'])

brewery_untapped_all
```

```python
brewery_apilist_df = pd.read_csv('brewery_list.csv') 
brewery_apilist_df = brewery_apilist_df.drop(columns=['Unnamed: 0','id','address_2', 'address_3','updated_at','created_at'])

#brewery_apilist_df.head()
```

```python
brewery_apilist_df.rename(columns = {'name':"brewery_name"},inplace = True)
brewery_apilist_df.head()
```

```python
brewery_final_df = brewery_apilist_df.merge(brewery_untapped_all, on='brewery_name', how='outer')
```

```python
brewery_final_df
```

```python
# brewery_final_df = brewery_final_df.drop_duplicates()
cols = ['brewery_type_x','brewery_type_y',]

brewery_final_df['brewery_type'] = brewery_final_df[cols].sum(1)
brewery_final_df.drop(cols, 1)
brewery_final_df
```

```python
brewery_final_df = brewery_final_df.drop_duplicates()
```

```python
brewery_final_df.to_csv("brewery_final_df.csv")
```
